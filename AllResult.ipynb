```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 전체 비교

excel_file_paths = ['LSTM_max_cumulative_returns.xlsx', 'LSTM_min_cumulative_returns.xlsx', 'GRU_max_cumulative_returns.xlsx', 'GRU_min_cumulative_returns.xlsx',
                    'ZeroNon_LSTM_max_cumulative_returns.xlsx', 'ZeroNon_LSTM_min_cumulative_returns.xlsx', 'ZeroNon_GRU_max_cumulative_returns.xlsx', 'ZeroNon_GRU_min_cumulative_returns.xlsx',
                    'ZeroZero_LSTM_max_cumulative_returns.xlsx', 'ZeroZero_LSTM_min_cumulative_returns.xlsx', 'ZeroZero_GRU_max_cumulative_returns.xlsx', 'ZeroZero_GRU_min_cumulative_returns.xlsx']

plot_name = ['LSTM_Original_Max', 'LSTM_Original_Min', 'GRU_Original_Max', 'GRU_Original_Min',
             'LSTM_NoInv_Max', 'LSTM_NoInv_Min', 'GRU_NoInv_Max', 'GRU_NoInv_Min',
             'LSTM_ZeroInv_Max', 'LSTM_ZeroInv_Min', 'GRU_ZeroInv_Max', 'GRU_ZeroInv_Min']

color_order = ['red', 'limegreen', 'chocolate', 'purple','tab:red', 'tab:cyan', 'gray', 'tab:blue', 'gold', 'hotpink', 'olive', 'navy']  # 부드럽고 편안한 색상을 지정합니다.

plt.rcParams['font.size'] = 20
plt.figure(figsize=(15, 7))

# 각 엑셀 파일을 순회하며 데이터를 읽고, plot
for i, excel_file_path in enumerate(excel_file_paths):
    df = pd.read_excel(excel_file_path)
    df = df.transpose()  # 데이터 전치
    plt.plot(df.loc[0], label=plot_name[i], color=color_order[i], linewidth=0.5)

plt.grid(True)
plt.xlabel('Investment Period')
plt.ylabel('Cumulative Log Returns')
plt.title('Cumulative Log Returns over Time (All)')
plt.legend()
plt.show()
```

```python
excel_file_paths = ['model_predictions_logreturns.xlsx', 'GRU_predictions_logreturns.xlsx']

plot_name = ['LSTM', 'GRU']

stocks = {
        'Tele' : ['VZ', 'DIS', 'T', 'CMCSA', 'EA'],
        'COND' : ['AMZN', 'HD', 'MCD', 'NKE', 'SBUX'],
        'ENRG': ['XOM', 'CVX', 'COP', 'EOG', 'SLB'],
        'HLCA' : ['JNJ', 'PFE', 'UNH','MRK','ABT'],
        'INDS' : ['BA', 'MMM','UNP','HON','RTX'],
        'MTRS': ['PPG', 'APD', 'ECL', 'SHW', 'NEM'],
        'FINC' : ['JPM', 'BRK-B', 'BAC', 'WFC', 'C'],
        'CONS' : ['PG', 'KO', 'PEP','WMT','MO'],
        'INFT' : ['AAPL','MSFT','INTC','CSCO','ORCL'],
        'UTIL' : ['NEE','DUK','D','SO','EXC'],
        'REES' : ['AMT', 'SPG', 'CCI', 'PLD', 'PSA']
    }


# 각 주식에 대한 그래프 그리기 위해 회사 리스트 생성
company_list = []
for sector in stocks:
    company_list += stocks[sector]

plt.rcParams['font.size'] = 20

plt.figure(figsize=(15, 7))

# 각 엑셀 파일을 순회하며 데이터를 읽고, plot
df1 = pd.read_excel(excel_file_paths[0])
df2 = pd.read_excel(excel_file_paths[1])
cum_df1 = np.array(np.cumsum(df1[174:], axis = 0))
cum_df2 = np.array(np.cumsum(df2[174:], axis = 0))


# 각 주식 별 그래프 그리기
for i, company in enumerate(company_list):
    plt.figure(figsize=(20,10))

    # LSTM 데이터 그래프
    plt.plot(cum_df1[:,i], color='red', label='LSTM ' + company)
    # GRU 데이터 그래프
    plt.plot(cum_df2[:,i], color='blue', label='GRU ' + company)

    plt.title('Cumulative log return of ' + company)
    plt.xlabel('Time')
    plt.ylabel('Cumulative Log Returns', fontsize=36)
    plt.legend()
    plt.show()
```

```python
plt.rcParams['font.size'] = 20

for sector in stocks:
    plt.figure(figsize=(20,10))

    # 실제 데이터 그래프
    avg_cum_df1 = np.mean([cum_df1[:, company_list.index(company)] for company in stocks[sector]], axis=0)
    plt.plot(avg_cum_df1, color='red', label='LSTM' + sector)

    # 예측 데이터 그래프
    avg_cum_df2 = np.mean([cum_df2[:, company_list.index(company)] for company in stocks[sector]], axis=0)
    plt.plot(avg_cum_df2, color='blue', label='GRU ' + sector)

    plt.title('Average Cumulative log return of ' + sector, fontsize=36)
    plt.xlabel('Time', fontsize=36)
    plt.ylabel('Cumulative Log Returns', fontsize=36)
    plt.legend()
    plt.show()
```

```python

plt.rcParams['font.size'] = 20

plt.figure(figsize=(20,10))

# 실제 데이터 그래프
allAvg_df1 = np.mean(cum_df1, axis=1)
plt.plot(allAvg_df1, color='red', label='Average Cumulative log-return of LSTM')

# 예측 데이터 그래프
allAvg_df2 = np.mean(cum_df2, axis=1)
plt.plot(allAvg_df2, color='blue', label='Average Cumulative log-return of GRU')

plt.title('Average Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Cumulative Log Returns', fontsize=36)
plt.legend()
plt.show()
```
